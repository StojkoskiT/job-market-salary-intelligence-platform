```python
import pandas as pd 
import matplotlib.pyplot as plt

df = pd.read_csv("../data/raw/ds_salaries.csv")
```

### Define skill keywords

```python
skills = [
    "python", "sql", "excel", "tableau", "power bi",
    "machine learnig", "deep learning", "nlp", "aws",
    "azure", "gcp", "spark", "hadoop"
]
```

### Exratct skills from job titles

```python
def extract_skills(title):
    title = title.lower()
    found = []

    for skill in skills:
        if skill in title:
            found.append(skill)

    return found

df["skills_found"] = df["job_title"].apply(extract_skills)
```

### Count skills

```python
from collections import Counter

all_skills = df["skills_found"].sum()

skill_counts = Counter(all_skills)

skill_df = pd.DataFrame(skill_counts.items(), columns=["Skill", "Count"])
skill_df = skill_df.sort_values(by="Count", ascending=False)

skill_df
```

### Plot top skills

```python
top_skills = skill_df.head(10)

top_skills.plot(kind="barh", x="Skill", y="Count", figsize=(10, 6))

plt.title("Most In-Demand Skills")
plt.xlabel("Frequency")
plt.gca().invert_yaxis()
plt.show()
```

Python and SQL dominate the data job market, appearing most frequently across job titles. Cloud technologies and big data tools are also increasingly in demand, indicating a shift toward scalable data infrastructure.
