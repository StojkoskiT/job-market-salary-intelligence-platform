### Salary by Job Title

```python
top_jobs = df.groupby("job_title")["salary_in_usd"].mean().sort_values(ascending=False).head(10)

top_jobs.plot(kind="barh", figsize=(10, 6))
plt.title("Top Playing Data Jobs")
plt.xlabel("Average Salary (USD)")
plt.gca().invert_yaxis()
plt.show()
```

Machine Learning Engineers and Data Scientists tend to have the highest average salaries, reflecting the high demand for advanced modeling and engineering skills.

### Salary by Country 

```python
top_countries = df.groupby("company_location")["salary_in_usd"].mean().sort_values(ascending=False).head(10)

top_countries.plot(kind="barh", figsize=(10, 6))
plt.title("Top Paying Countires")
plt.xlabel("Average Salary (USD)")
plt.gca().invert_yaxis()
plt.show()
```

Countries like the US and Switzerland dominate salary rankings, indicating strong demand and higher compensation standards in developed markets.

### Remote Work Analysis

```python
df["remote_ratio"].value_counts().plot(kind="bar")

plt.title("Remote Work Distribution")
plt.xlabel("Remote Ratio")
plt.ylabel("Count")
plt.show()
```

A significant portion of data roles offer remote or hybrid flexibility, highlighting the global and digital nature of the data industry.

### Experience vs Salary

```python
df.groupby("experience_level")["salary_in_usd"].mean().plot(kind="bar")

plt.title("Salary by Experience Level")
plt.xlabel("Experience Level")
plt.ylabel("Average Salary")
plt.show()
```

Senior-level professionals earn significantly more than entry-level roles, emphasizing the value of experience in the data field.
